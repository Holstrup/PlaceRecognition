```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import imageio
import numpy as np
import scipy.stats as stats
```

```python
anchor = imageio.imread('../data/data/train_val/boston/database/images/__c5lJhqEW3eRiRpupMkbA.jpg')
positive = imageio.imread('../data/data/train_val/boston/database/images/JZ8WpWAF08WWghrqp1VR6A.jpg')
negative = imageio.imread('../data/data/train_val/boston/database/images/__q_C1ayZ-HyG8R6LdNlOA.jpg')
images = [anchor, positive, negative]
titles = ["anchor", "positive", "negative"]
```

```python
w=10
h=10
fig=plt.figure(figsize=(20, 20))
columns = 3
rows = 1

for i, image in enumerate(images):
    fig.add_subplot(rows, columns, i + 1)
    plt.axis('off')
    plt.title(titles[i])
    plt.imshow(image)
    
plt.show()
```

```python
w=10
h=5
fig=plt.figure(figsize=(20, 5))
columns = 3
rows = 1

for i, image in enumerate(images):
    fig.add_subplot(rows, columns, i + 1)
    plt.axis('off')
    plt.title(titles[i])
    plt.hist(image.ravel(),256,[0,256])
    
plt.show()
```
